```python
import torchvision
```

```python
mnist = torchvision.datasets.MNIST(root="", train=True, download=True)
```

```python
x = mnist.data.numpy()
y = mnist.targets.numpy()
```

```python
data = []

digits = [2, 3, 4]

for digit in digits:
    mask = y == digit
    data.append(x[mask][:200])

data = np.concatenate(data, axis=0)
```

```python
# shuffle data
np.random.seed(0)
np.random.shuffle(data)
data[0]
```

```python
_, ax = plt.subplots(1, 1, figsize=(12, 12))
ax.imshow(data[0])
ax.axis('off')
plt.show()
```

```python
data = (data > 127.0).astype(np.float32)
```

```python
data[0]
```

```python
_, ax = plt.subplots(1, 1, figsize=(12, 12))
ax.imshow(data[0])
ax.axis('off')
plt.show()
```

```python
n_row = 20
n_col = 30

_, axs = plt.subplots(n_row, n_col, figsize=(12, 12))
axs = axs.flatten()
for img, ax in zip(data, axs):
    ax.imshow(img)
    ax.axis('off')
plt.show()
```
